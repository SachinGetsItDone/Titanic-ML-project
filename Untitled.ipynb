## Data Cleaning 

```python
print("Missing values>> ")
df = data.copy()
df.isna().sum()
```

```python
df["Age"] = df["Age"].median()
```

- Filled age with median
- Age distribution can be skewed, we use median since it is not affected by outliers and more accurately represents the most probable age of the passenger

```python
df["Cabin"].unique()
```

# Since cabin values vary widely, we will assign 0 for no cabin and 1 for having a cabin.

```python
df["Has_Cabin"] = df['Cabin'].notna().astype(int)
```

```python
df.drop(["Cabin"], axis = 1, inplace = True)
```

## Embarked can filled with mode values 

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Verify that there are no more missing values in the columns we handled so far
print("Missing values after Embarked cleaning:")
print(df[['Age', 'Embarked']].isna().sum())
```

# Feature Engineering

```python
df.head()
```

## Name, PassengerID, Ticket are of no use for predicting surviability 
- drop these 3
- merge sibsp parch + 1

```python
df.drop(['Name', 'PassengerId', 'Ticket'],axis = 1,  inplace = True)
```

```python
df['Familiy'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['Familiy'].unique()
```

```python
df['Embarked'].unique() #Can do label encoding for this section
```

```python
le = LabelEncoder()
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
#binary encoding this section 
df['Sex'] = df['Sex'].map({
    'female': 0,
    'male': 1
})
```

```python
df.head()
```

```python
df.drop(['SibSp', 'Parch'], axis = 1, inplace = True)
```

```python
df.head()
df.rename(columns={'Familiy': 'Family'}, inplace=True)
```
